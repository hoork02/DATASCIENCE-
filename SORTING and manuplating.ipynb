## Data Cleaning: Handling Missing Data

Usually, before we can process or analyze our data further, we need to clean it to get more accurate and meaningful results.

Data cleaning is the process of preparing data for analysis by removing or modifying data that is incorrect, incomplete, irrelevant, duplicated, or improperly formatted. In this part of the tutorial, we will be looking at how you can handle missing values/incomplete data.

Missing values are typically represented by NaN or NULL values in your dataset. You should start by checking for the presence of such values.

```python
#first check if there are any NULL values in any of the rows

df.isnull().any()

#for this Iris df, we should get False for each of the columns. This means that for each of these columns, there is no row with a NULL value.
```

```python
#Let's try another dataset. There are some NULL values in this one.
df2 = pd.read_csv('/content/drive/MyDrive/pandas_lab/Iris2.csv')

print(df2.shape)

#df2.isnull().any()
df2.isnull().sum()
```

Depending on the datatype of the values in your column (Integer or String) and the number of missing values, you can choose how to deal with missing values. Some commonly used methods are:



*   .dropna(): drop any rows with missing values for the entire dataset or any given column. See the examples in the notebook.

*   .fillna(): replace the missing values of column with a constant or the column mean, median, mode, etc.

```python
#There are several ways we can handle missing values. The simplest is to drop any rows with missing values for the entire dataset or any given column

df3 = df2.dropna()
```

```python
print(df3.shape)
df3.isnull().sum()
```

```python
#We can drop NULL values for just one column too

df3 = df2.dropna(subset = ['Species'])
```

```python
print(df3.shape)
df3.isnull().sum()
```
